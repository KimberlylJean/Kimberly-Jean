```python
import pandas as pd
import numpy as np
import string
import re
import nltk
from nltk.corpus import stopwords,wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import jaccard_score
import operator
import collections
from rake_nltk import Rake

nltk.download('wordnet')

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

# Importing the cleaned data
movies2 = pd.read_csv(r"C:\Users\kimyj\OneDrive\Documents\MSBA_Spring23_Documents\CIS9660\GroupProject\movie2",index_col=0)
movies2.head(4)

# Standarizing the popularity column
scaler = StandardScaler()
movies2['popularity_standardized'] = scaler.fit_transform(movies2[['popularity']])
movies2['release_year'] = scaler.fit_transform(movies2[['release_year']])

# Extracting the important words from the overview column
def extract_key_words(input_str, top_n_keywords=5):
    r = Rake()
    r.extract_keywords_from_text(input_str.lower())
    key_words_dict_scores = r.get_word_degrees()
    sorted_key_words_dict_scores = sorted(key_words_dict_scores.items(), key=operator.itemgetter(1), reverse=True)
    sorted_dict = collections.OrderedDict(sorted_key_words_dict_scores)
    
    # Change top_n_keywords as appropriate
    return sorted_dict, list(sorted_dict.keys())[:top_n_keywords]

movies2['short_overview'] = movies2['overview'].apply(lambda x: extract_key_words(x)[1]).apply(' '.join)

movies2.head(4)

def clean_text_for_bigrams(text):
    text = re.sub(r'\s+', ' ', text) # remove extra spaces
    text = text.strip() # remove leading/trailing spaces
    text = text.lower() # convert text to lowercase
    return text

# Creating a function to clean the data 

def clean_text_for_td_idf(text):
    # Remove punctuation
    text = "".join([i for i in text if i not in string.punctuation])
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove stop words/tokenization
    stop_words = nltk.corpus.stopwords.words('english')
    ps = nltk.PorterStemmer()
    words = re.split('\W+', text) #tokenization
    text = [ps.stem(word) for word in words if word not in stop_words] #removing stop words
    
    text_2 = ' '.join(word for word in text)
    return text_2     # Returns one string

# Applying the functions to the columns
movies2['clean_overview'] = movies2['short_overview'].apply(clean_text_for_td_idf)
movies2['clean_genres'] = movies2['genres_list'].apply(clean_text_for_td_idf)
movies2['clean_keywords'] = movies2['keywords_list'].apply(clean_text_for_td_idf)
movies2['clean_directors'] = movies2['directors'].apply(clean_text_for_bigrams)
movies2['clean_characters'] = movies2['main_characters'].apply(clean_text_for_bigrams)

movies2.head(4)

# Create CountVectorizer with bigram range
bigram_vectorizer1 = CountVectorizer(ngram_range=(2, 2))
bigram_vectorizer2 = CountVectorizer(ngram_range=(2, 2))

# Fit and transform the director and main_characters columns
directors_bigram = bigram_vectorizer1.fit_transform(movies2['clean_directors'])
characters_bigram = bigram_vectorizer2.fit_transform(movies2['clean_characters'])

# Vectorizing the data
tfidf_vect1 = TfidfVectorizer(analyzer=clean_text_for_td_idf)
tfidf_vect2 = TfidfVectorizer(analyzer=clean_text_for_td_idf)
tfidf_vect3 = TfidfVectorizer(analyzer=clean_text_for_td_idf)

overview_matrix = tfidf_vect1.fit_transform(movies2['clean_overview'])
genres_matrix = tfidf_vect2.fit_transform(movies2['clean_genres'])
keywords_matrix = tfidf_vect3.fit_transform(movies2['clean_keywords'])



# Transforming the matrices to dataframes
overview_matrix = pd.DataFrame(overview_matrix.toarray(), columns=tfidf_vect1.get_feature_names_out())
genres_matrix = pd.DataFrame(genres_matrix.toarray(), columns=tfidf_vect2.get_feature_names_out())
keywords_matrix = pd.DataFrame(keywords_matrix.toarray(), columns=tfidf_vect3.get_feature_names_out())

directors_bigram = pd.DataFrame(directors_bigram.toarray(), columns=bigram_vectorizer1.get_feature_names_out())
characters_bigram = pd.DataFrame(characters_bigram.toarray(), columns=bigram_vectorizer2.get_feature_names_out())

%%time

# Calculate the similarity matrix for the overview feature
overview_sim_matrix = cosine_similarity(overview_matrix)

# Calculate the similarity matrix for the genres feature
genres_sim_matrix = cosine_similarity(genres_matrix)

# Calculate the similarity matrix for the keywords feature
keywords_sim_matrix = cosine_similarity(keywords_matrix)

# Calculate the similarity matrix for the directors feature
directors_sim_matrix = cosine_similarity(directors_bigram)

# Calculate the similarity matrix for the characters feature
characters_sim_matrix = cosine_similarity(characters_bigram)

# Calculate the similarity matrix for the popularity feature
popularity_corr_matrix = np.corrcoef(movies2['popularity_standardized'])


# Calculate the similarity mtrix for the year of release
year_corr_matrix = np.corrcoef(movies2['release_year'])

print("overview_sim_matrix shape:", overview_sim_matrix.shape)
print("genres_sim_matrix shape:", genres_sim_matrix.shape)
print("keywords_sim_matrix shape:", keywords_sim_matrix.shape)
print("directors_sim_matrix shape:", directors_sim_matrix.shape)
print("characters_sim_matrix shape:", characters_sim_matrix.shape)
print("popularity_sim_matrix shape:", popularity_corr_matrix.shape)
print("year_sim_matrix shape:", year_corr_matrix.shape)


#cos_similarity_matrix = (overview_sim_matrix + genres_sim_matrix + keywords_sim_matrix + directors_sim_matrix + characters_sim_matrix) / 5
#overall_sim_matrix = 0.1 * year_corr_matrix + 0.2 * popularity_corr_matrix+ 0.7 * cos_similarity_matrix


# Define the weights for each feature
weights = {
    'overview': 0.3,
    'genres': 0.25,
    'keywords': 0.15,
    'directors': 0.1,
    'characters': 0.1,
    'popularity': 0.05,
    'release_year': 0.05
}

# Compute the overall similarity matrix as a weighted average of the individual similarity matrices
overall_sim_matrix = (
    weights['overview'] * overview_sim_matrix +
    weights['genres'] * genres_sim_matrix +
    weights['keywords'] * keywords_sim_matrix +
    weights['directors'] * directors_sim_matrix +
    weights['characters'] * characters_sim_matrix +
    weights['popularity'] * popularity_corr_matrix +
    weights['release_year'] * year_corr_matrix
) / sum(weights.values())


%%time

similarities = {}
for i in range(len(overall_sim_matrix)):
    # Sort each element in cosine_similarities and get the indexes of the 50 most similar songs
    similar_indices = overall_sim_matrix[i].argsort()[:50:-1]
    
    # Exclude the first element (i.e., the same movie) from the list of similar movies
    similar_indices = similar_indices[1:]
    
    # Store in similarities each name of the 50 most similar songs
    similarities[movies2['title_y'].iloc[i]] = [(overall_sim_matrix[i][x], movies2['title_y'][x], movies2['genres_list'][x],movies2['directors'][x],movies2['clean_characters'][x] )
                                               for x in similar_indices]


class ContentBasedRecommender:
    def __init__(self, matrix):
        self.matrix_similar = matrix

    def _print_movies(self, title, recommended_movies):
        rec_items = len(recommended_movies)
        
        print(f'The {rec_items} recommended movies for {title} are:')
        for i in range(rec_items):
            print(f"\t{i+1}: {recommended_movies[i][1].title()} \n The genres of this movie are: {recommended_movies[i][2].title()}.\n The director is :{recommended_movies[i][3].title()}.\n The lead actors are :{recommended_movies[i][4].title()}.\n (Similarity score : {round(recommended_movies[i][0]*100, 3)}%)\n")

        
    def recommend(self, title, rec_num_movies=5):
        title = title.lower().strip()
        try:
            recommended_movies = self.matrix_similar[title][:rec_num_movies]
            # Print each recommended movie
            self._print_movies(title=title, recommended_movies=recommended_movies)
        except KeyError:
            print(f"Movie '{title}' not found in dataset.")


# Instantiate Content based recommender class with the precomputed similarities
recommendations = ContentBasedRecommender(similarities)

#recommendations.recommend('Frozen')

#recommendations.recommend('The Shawshank Redemption')

#recommendations.recommend('The Dark Knight')

#recommendations.recommend('The Avengers')

import tkinter as tk
from tkinter import scrolledtext

class ContentBasedRecommenderGUI:
    def __init__(self, matrix):
        self.matrix_similar = matrix

        # Create the main window
        self.root = tk.Tk()
        self.root.title("Movie Recommender Platform")

        # Create the input label and text box
        input_frame = tk.Frame(self.root)
        input_label = tk.Label(input_frame, text="Enter a movie title:")
        input_label.pack(side="left")
        self.input_text = tk.Entry(input_frame, width=30)
        self.input_text.pack(side="left")
        input_frame.pack()

        # Create the recommendation button
        button_frame = tk.Frame(self.root)
        self.recommend_button = tk.Button(button_frame, text="Recommend", command=self.recommend_movies)
        self.recommend_button.pack()
        button_frame.pack()

        # Create the output label and text widget
        output_label = tk.Label(self.root, text="Recommended movies:")
        output_label.pack()
        self.output_text = scrolledtext.ScrolledText(self.root, width=100, height=30)
        self.output_text.pack()

    def run(self):
        self.root.mainloop()

    def _print_movies(self, title, recommended_movies):
        rec_items = len(recommended_movies)
        
        output = f'The {rec_items} recommended movies for {title} are:\n'
        for i in range(rec_items):
            output += f"\t{i+1}: {recommended_movies[i][1].title()} \n The genres of this movie are: {recommended_movies[i][2].title()}.\n The director is :{recommended_movies[i][3].title()}.\n The lead actors are :{recommended_movies[i][4].title()}.\n (Similarity score : {round(recommended_movies[i][0]*100, 3)}%)\n"

        
        self.output_text.delete(1.0, tk.END)
        self.output_text.insert(tk.END, output)

    def recommend_movies(self):
        title = self.input_text.get()
        title = title.lower().strip()
        recommended_movies = self.matrix_similar[title][:5]
        self._print_movies(title=title, recommended_movies=recommended_movies)


# Example usage
matrix = similarities  # replace with your similarity matrix
gui = ContentBasedRecommenderGUI(matrix)
gui.run()


```
